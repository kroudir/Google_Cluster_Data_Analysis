## Data preparation. RDD

For this part of questions we will be using PythonRDD.

Start Spark, prepare needed dataset, parce it and map into RDD.

```python
# Start spark with 1 worker thread
sc = SparkContext("local[*]")
sc.setLogLevel("ERROR")
```

```python
# Set dataset column names.
col = {
    "time" : 0,
    "missingInfo" : 1,
    "jobID" : 2,
    "taskID" : 3,
    "machineID" : 4,
    "eventType" : 5,
    "user" : 6,
    "schedulingClass" : 7,
    "priority" : 8,
    "cpuReq" : 9,
    "memoryReq" : 10,
    "diskSpaceReq" : 11,
    "dms" : 12 # different machines restriction
}

# Preparing database by splitting it and setting data type for each field
task_events_rdd = sc.textFile(path_task_events)
te_entries = task_events_rdd.map(lambda x : x.split(','))
te_entries = te_entries.map(lambda x : [
    int('0' + x[col["time"]]),
    int('0' + x[col["missingInfo"]]),
    int('0' + x[col["jobID"]]),
    int('0' + x[col["taskID"]]),
    int('0' + x[col["machineID"]]),
    int('0' + x[col["eventType"]]),
    str(x[col["user"]]),
    int('0' + x[col["schedulingClass"]]),
    int('0' + x[col["priority"]]),
    float('0' + x[col["cpuReq"]]),
    float('0' + x[col["memoryReq"]]),
    float('0' + x[col["diskSpaceReq"]]),
    bool(x[col["dms"]])
])

# Keep the RDD in memory
te_entries.cache()
```

## On average, how many tasks compose a job?

To calculate this value we need to go through the following steps:
- MapReduce data by **jobID** to get a correlation of **jobID** to list of tasks that compose it.
- MapReduce correlation about to get **jobID** to number of tasks that compose it.
- Calculate mean of number of tasks over all **jobID**s.


```python
# Create a new pair-based RDD.
# First element of which is 'key' (value of 'key' field in dataset),
# Second is corresponding line or a value of 'value' field in dataset if 'value' arg is passed.
def regroup_RDD_by_key(rdd, key, value=None):
    key_num = col[key]
    
    if value is not None:
        val_num =  col[value]
        return rdd.map(lambda line: (line[key_num], line[val_num]))
    
    return rdd.map(lambda line : (line[key_num], line))

# Calculate mean of 'num' (0 or 1) elements of dictionary of pairs
def mean_of_elems(data, num):
    sum_val = 0
    for elem in data:
        sum_val += elem[num]
    return sum_val/len(data)
```

```python
# MapReduce data by jobID to get a correlation of jobID to number of tasks that compose it.
jobID_to_taskIDnum = regroup_RDD_by_key(te_entries, 'jobID', 'taskID').countByKey()
# Calculate mean of numbers of tasks.
mean = mean_of_elems(jobID_to_taskIDnum.items(), 1)
print("In general", mean, "tasks compose one job.")
```

# What can you say about the relation between the scheduling class of a job, the scheduling class of its tasks, and their priority?

To get a relation between scheduling class and a priotity we calculate the mean of priority that is usually assigned to the scheduling class.

For this we divide calculation in two stages:
1. Gather all tasks of same scheduling class 'i'
2. Calculate mean of their priority

And to have better representation, build bar chart of relation between scheduling class and a mean of its 
priority.

### Scheduling class
Represents latency-sensitivity.
0 - the least sensitive
3 - the most sensitive

### Priority
Ranges from 0 (the lowest priority) to 11 (the highest priority).

```python
# Before we start let's make sure that any task's scheduling_class is the same as of its job.

# Join tables by job_id to compare scheduling_class'es.
df_job_task = df_job_event.withColumnRenamed('scheduling_class','job_scheduling_class')\
                .join(df_task_event.withColumnRenamed('scheduling_class','task_scheduling_class'),
                      'job_id', 'outer')

# Check that scheduling class of job and of it's task is the same.
assert(df_job_task.filter(df_job_task['job_scheduling_class'] != df_job_task['task_scheduling_class']).count() == 0)
```

```python
#Create a dataframe with Scheduling Class and a mean of it's priority.
df_sch_class_priority = df_task_event.groupBy("scheduling_class") \
        .agg(F.collect_list("priority").alias('priority_list'), \
             F.mean('priority').alias('average_priority'))

#df_sch_class_priority.show(5)
```

```python
# Build bar chart based on found values.

#Convert the spark dataframe into a Pandas a dataframe to be able to use seaborn
pd_df_sch_class_priority = df_sch_class_priority.toPandas().sort_values(by='scheduling_class')
ax = sns.barplot(x='scheduling_class',y = 'average_priority', data = pd_df_sch_class_priority)
ax.set(xlabel='Scheduling class', ylabel='Average priority', title = 'Distribution of average priority between scheduling classes')

#Adding the pourcentages on the Top of the bars
av_priority_list = pd_df_sch_class_priority['average_priority'].tolist()
i = 0
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2.,
            height,
            '{:1.2f}'.format(av_priority_list[i]),
            ha="center")
    i = i+1

plt.savefig('prior_sch_class.png', bbox_inches='tight')
plt.show()
```

# Do tasks with low priority have a higher probability of being evicted?

The task_events table contains field "event type" which can take values from 0 to 8. The one that interests us to answer this question is
   >EVICT(2): A task or job was descheduled because of a higher priority task or job,
because the scheduler overcommitted and the actual demand exceeded the machine
capacity, because the machine on which it was running became unusable (e.g. taken
offline for repairs), or because a disk holding the task’s data was lost.

To answer the queston we firstly calcutale correspondance of priority to number of times it was evicted. And then calculate what percentage this number makes of all times any task of any priority was evicted (sum of all lines with event_type = 2).

```python
# Select lines with event_type = 2.
df_evicted = df_task_event.filter(df_task_event['event_type'] == 2)

# Group by priority and count how many times each priority was evicted.
df_priority_evicted = df_evicted.groupBy("priority").count()

# Get common number of evicted events.
all_evicted = df_priority_evicted.select('count').groupBy().sum().collect()[0][0]

# Calculate chance of each priority to be evicted.
df_priority_eviction = df_priority_evicted.withColumn('chance_of_eviction', F.col('count')*100/all_evicted)

#df_priority_eviction.show()
```

```python
# Tasks with the lowest priorities get evicted almost 100% 
# which makes it impossible to observe a distribution among other priorities.

# Create second dataframe without the lowest priorities.
df_priority_eviction_without_zero = df_priority_eviction.filter(F.col('priority') > 1)
df_priority_eviction_without_zero.show()
```

```python
# Build two graphics - with all priorities and without the lowest to observe the distribution.
pd_df_priority_eviction = df_priority_eviction.toPandas()
pd_df_priority_eviction_without_zero = df_priority_eviction_without_zero.toPandas()

_, axes = plt.subplots(1, 2, sharey=False)

ax0 = sns.lineplot(x='priority',y = 'chance_of_eviction', data = pd_df_priority_eviction, ax = axes[0])
ax1 = sns.lineplot(x='priority',y = 'chance_of_eviction', data = pd_df_priority_eviction_without_zero, ax = axes[1])

ax0.set(xlabel='Priority', ylabel='Chance of eviction', title = 'Disctribution of chances of eviction \n for all priorities.')
ax1.set(xlabel='Priority', ylabel='Chance of eviction', title = 'Disctribution of chances of eviction \n for priorities from 2 to 10.')

plt.savefig('evict_priority.png', bbox_inches='tight')
plt.show()
```

# In general, do tasks from the same job run on the same machine?

To answer this question we need only task_event table.
The table contains **job_id** and **machine_id** fields. Thus it is sufficient to gather all **machine_id**'s for the same **job_id** to see if there's multiple machines for tasks of the same job.

```python
df_job_machines = df_task_event.groupBy('job_id') \
                    .agg(F.countDistinct("machine_id").alias('number_of_machines'))

#df_job_machines.show()
```

```python
# To answer the question we calculate a percentage of how many jobs run on the same machine
# and how many run on multiple machines.

run_on_one_machine = df_job_machines.filter(df_job_machines['number_of_machines'] == 1).count()
run_on_multiple_machines = df_job_machines.filter(df_job_machines['number_of_machines'] != 1).count()

percentage_on_one_machine = run_on_one_machine * 100/(run_on_one_machine + run_on_multiple_machines)
percentage_on_multiple_machine = run_on_multiple_machines * 100/(run_on_one_machine + run_on_multiple_machines)
```

```python
# Draw pie chart of percentage of jobs running on one machine to percentage of jobs running on multiple machines.
labels = 'Run on the same machine', 'Run on multiple machines'
sizes = [percentage_on_one_machine, percentage_on_multiple_machine]
explode = (0.1, 0)
colors = ['gold', 'lightcoral']
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
    autopct='%1.1f%%', shadow=True)
plt.title('Distribution of jobs that run on the same machine and jobs that run on multiple machines')

plt.savefig('same_multi_maching_job.png', bbox_inches='tight')
plt.show()
```

# Is there a relation between the amount of resource consumed by tasks and their priority?

To answer this question we need data from task_usage table, particularly fields like memory usage, disk space usage, CPU usage. We will try to see a relation between each type of resources consumed to priority.

## Prepare dataframe

## Schema
The task resource usage table contains these fields:
1. start time of the measurement period
2. end time of the measurement period
3. job ID
4. task index
5. machine ID
6. mean CPU usage rate
7. canonical memory usage
8. assigned memory usage
9. unmapped page cache memory usage
10.total page cache memory usage
11. maximum memory usage
12. mean disk I/O time
13. mean local disk space used
14. maximum CPU usage
15. maximum disk IO time
16. cycles per instruction (CPI)
17. memory accesses per instruction (MAI)
18. sample portion
19. aggregation type (1 if maximums from subcontainers were summed)
20. sampled CPU usage

```python
#Define a schema for the task_usage table
schema_task_events = StructType([
    StructField("start_time", LongType(), False),
    StructField("end_time", LongType(), False),
    StructField("job_id", LongType(), False),
    StructField("task_index", IntegerType(), False),
    StructField("machine_id", LongType(), False),

    StructField("cpu_rate", FloatType(), True),
    StructField("canonical_memory_usage", FloatType(), True),
    StructField("assigned_memory_usage", FloatType(), True),
    StructField("unmapped_page_cache", FloatType(), True),
    StructField("total_page_cache", FloatType(), True),
    StructField("maximum_memory_usage", FloatType(), True),

    StructField("disk_io_time", FloatType(), True),
    StructField("local_disk_space_usage", FloatType(), True),
    StructField("max_cpu_rate", FloatType(), True),
    StructField("max_disk_io_time", FloatType(), True),

    StructField("cycles_per_instruction", FloatType(), True),
    StructField("memory_access_per_instruction", FloatType(), True),
    StructField("sample_portion", FloatType(), True),
    StructField("aggregation_type", BooleanType(), True),
    StructField("sampled_cpu_usage", FloatType(), True)])

#Read the data into a SparkDataFrame
df_task_usage = spark.read.csv(path_task_usage, header = False, schema = schema_task_events)
```

```python
# Select only data from task_usage that we'll be working with:
df_task_only_usage = df_task_usage.select(['job_id', 'task_index', 'canonical_memory_usage', 'assigned_memory_usage', 
                      'maximum_memory_usage', 'local_disk_space_usage', 'sampled_cpu_usage'])

# Select important for the answer columns from task_event: task_index and priority columns from df_task_events.
df_task_id_piority = df_task_event.select(['job_id','task_index', 'priority'])

# Join priority column to df_task_usage table based on values of job_id and task_index.
# And pick only resources and priority rows.
df_resource_usage_priority = df_task_only_usage.join(df_task_id_piority, 
                            (df_task_id_piority['job_id'] == df_task_only_usage['job_id']) & 
                            (df_task_id_piority['task_index'] == df_task_only_usage['task_index']))\
                            .select(['canonical_memory_usage', 'assigned_memory_usage', 'maximum_memory_usage', 
                                     'local_disk_space_usage', 'sampled_cpu_usage', 'priority'])

# Now we can calculate average of consumation of each resource for each priority.
priority_mean_usage = df_resource_usage_priority.groupBy('priority')\
                     .agg(F.mean('canonical_memory_usage').alias('canonical_memory_usage_mean'),
                          F.mean('assigned_memory_usage').alias('assigned_memory_usage_mean'),
                          F.mean('maximum_memory_usage').alias('maximum_memory_usage_mean'),
                          F.mean('local_disk_space_usage').alias('local_disk_space_usage_mean'),
                          F.mean('sampled_cpu_usage').alias('sampled_cpu_usage_mean'))
```

Build graphs to better explore relationship between priority and resource usage.

```python
pd_priority_mean_usage = priority_mean_usage.toPandas()
```

```python
# Explore relationship between priority and canonical memory usage.
ax = sns.barplot(x='priority',y = 'canonical_memory_usage_mean', data = pd_priority_mean_usage)
ax.set(xlabel='Priority', ylabel='Mean of Canonical memory usage', title = 'Disctribution of canonical memory usage\n according to priority of a task')

plt.savefig('canon_mem_prior.png', bbox_inches='tight')
plt.show()
```
