## Data Presentation
In this part of the analysis, we will be using the machine events table
### Machine events
Each machine is described by one or more records in the machine event table. 
#### Schema:
1. timestamp
2. machineID
3. eventtype
4. platformID
5. capacity:CPU
6. capacity:memory
> There are three types of machine events:
>- ADD (0): a machine became available to the cluster - all machines in the trace will
have an ADD event.
>- REMOVE (1): a machine was removed from the cluster. Removals can occur due to
failures or maintenance.
>- UPDATE (2): a machine available to the cluster had its available resources changed.

### 2. What is the poucentage of computational time lost due to maintenance (a machine went offline and reconnected later)? What is the distribution of this time over CPU capacities/ Memory Capacities ?

```python
def calculate_down_time(event_type_list, time_list,max_active_time):
    """
    Calculate the pourcentage down time of a machine
    Arguments : 
        event_type_list : a list of the events
        time_list : a list of the timestamps of occurrence of each event
    """
    
    full_down_time = 0
    
    #Loop over the list of events
    for i in range(len(event_type_list) - 1):
        if event_type_list[i] == 1 and event_type_list[i + 1] == 0: #if current event is DELETE and the coming event is ADD
            down_time = time_list[i + 1] - time_list[i] #Calculate the down time
            full_down_time += down_time 
    
    #if the last event is DELETE
    if event_type_list[len(event_type_list) - 1] == 1:
        down_time = max_active_time - time_list[len(event_type_list) - 1] #Calculate the down time
        full_down_time += down_time 
    
    real_max_active_time = max_active_time - time_list[0]
    print(real_max_active_time,full_down_time)
    if ( real_max_active_time > 0):
        return float(full_down_time / real_max_active_time) * 100
    
    return 0.0

#Create a spark UDF to calculate the down time
calculate_down_time_udf = F.udf(lambda event_type_list, time_list, max_active_time: calculate_down_time(event_type_list,time_list,max_active_time), FloatType())
```

```python
#Create a dataframe where each line is a unique machine and has events and time as a list as separate columns
df_machine_events_lists = df_machine_event.groupBy('machine_id') \
                    .agg(F.collect_list("event_type").alias('event_type_list') \
                         ,F.collect_list("time").alias('time_list') \
                         ,F.mean('cpu_capacity').alias('cpu_capacity') \
                         ,F.mean('memory_capacity').alias('memory_capacity'))

#check that the dataframe has all the machines
assert df_machine_events_lists.count() == num_unique_machines

#Get the end time, which is the maximum in the time column is the original dataset
time_end = df_machine_event.agg({"time": "max"}).collect()[0][0]

#Add the maximum active time column
df_machine_events_lists = df_machine_events_lists.withColumn('max_active_time', F.lit(time_end))

#Create a dataframe with a new column down_time that has the down time of the specefied machine.
df_machine_events_down_time = df_machine_events_lists.select('machine_id',
                                                'cpu_capacity',
                                                'memory_capacity',
                                                'event_type_list',
                                                'time_list',
                                                'max_active_time',
                    calculate_down_time_udf('event_type_list','time_list','max_active_time').alias('down_time_percentage'))

#Create active time and pourcentage down time columns
df_machine_events_down_active_time = df_machine_events_down_time.select('machine_id',
                                        'cpu_capacity','memory_capacity', 'down_time_percentage')

df_machine_events_down_active_time.show(10)
```

### 3. What is the percentage of computational power lost due to maintenance (a machine went offline and reconnected later)? What is the distribution of this lost over CPU capacities/ Memory Capacities ?

```python
total_capacity = df_machine_events_down_active_time.agg(F.sum('cpu_capacity')).collect()[0][0]
print('The total calculation capacity in our cluster is : {}'.format(total_capacity))
```

```python
#Create lost_power_pourcentage column
df_machine_events_lost_power_no_anomaly = \
        df_machine_events_down_active_time \
        .withColumn('lost_power_percentage',
                    F.col('down_time_percentage') * F.col('cpu_capacity'))

df_machine_events_lost_power_no_anomaly.select('machine_id','down_time_percentage','cpu_capacity','lost_power_percentage');
```

```python
average_percentage_of_lost_power = df_machine_events_lost_power_no_anomaly.agg(F.mean('lost_power_percentage')).collect()[0][0]
print('The average lost of power of computation lost to Maintenance is {:1.2f}%'.format(average_percentage_of_lost_power))
```
